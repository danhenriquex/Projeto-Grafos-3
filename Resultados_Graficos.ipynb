```python
import matplotlib.pyplot as plt

def show_results(file1: str, file2: str):
  x = []
  y_random = []
  y_karge = []

  with open(file1) as fp:
    line = fp.readline().split()
    while line:
        x.append(line[0])
        y_random.append(line[1])
        line = fp.readline().split()

  with open(file2) as fp:
    line = fp.readline().split()
    while line:
        y_karge.append(line[1])
        line = fp.readline().split()

  fig, ax = plt.subplots()

  ax.plot(x, y_karge,label = 'Karge')
  ax.plot(x, y_random,label = 'Random')
  ax.legend(loc = 'upper left')
  plt.ylabel('Prob')
  plt.xlabel('MaxInt')
  plt.title('Grafo Type2_2')
  plt.show()
```

# Arquivo 1.1


```python

show_results("./resultados/resultados_type1_1_karge.txt", "./resultados/resultados_type1_1_random.txt")
```
